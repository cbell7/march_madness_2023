# Initial output

## Bring in team names and spot check

```python
teams_m = pd.read_csv('data/MTeams.csv')
teams_w = pd.read_csv('data/WTeams.csv')
```

```python
teams_m.head()
```

```python
teams_m = teams_m[["TeamID", "TeamName"]]
teams_w = teams_w[["TeamID", "TeamName"]]
```

```python
pred_df_m.index.to_series().str.split("_", expand=True)
```

```python
# Get team and opponent IDS
pred_df_m[["Season", "TeamID", "OppTeamID"]] = pred_df_m.index.to_series().str.split("_", expand=True).astype('int64')
pred_df_w[["Season", "TeamID", "OppTeamID"]] = pred_df_w.index.to_series().str.split("_", expand=True).astype('int64')
```

```python
team_preds_m = (pred_df_m
 .merge(teams_m)
 .merge(teams_m.rename(columns={"TeamID": "OppTeamID", "TeamName": "OppName"}))
 .filter(["Seed", "TeamName", "OppSeed", "OppName", "KPDiff", "Pred"]))
```

```python
team_preds_m
```

```python
team_preds_w = (pred_df_w
 .merge(teams_w)
 .merge(teams_w.rename(columns={"TeamID": "OppTeamID", "TeamName": "OppName"}))
 .filter(["Seed", "TeamName", "OppSeed", "OppName", "KPDiff", "Pred"]))
```

```python
team_preds_w
```

```python
# Write to csv
team_preds_m.to_csv('archive/team_preds_m2.csv', index=False)
```

# Build final submission file

```python
submission_file = pd.read_csv('data/SampleSubmission2023.csv')
```

```python
submission_file
```

```python
pred_df_m_final = pred_df_m.reset_index().filter(["ID", "Pred"])
```

```python
pred_df_m_final.dtypes
```

```python
submission_file.dtypes
```

```python
pred_df_w_final = pred_df_w.reset_index().filter(["ID", "Pred"])
```

```python
pred_df_w_final.dtypes
```

```python
# Combine dataframes
pred_df_combined = pd.concat([pred_df_m_final, pred_df_w_final]).reset_index(drop=True)
```

```python
pred_df_combined
```

```python
# Merge predictions with sample file
final_output = (submission_file
 .drop(columns="Pred")
 .merge(pred_df_combined, how="left")
 .fillna(0.5)
)
```
